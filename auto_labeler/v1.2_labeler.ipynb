```python
import csv
from glob import glob

from PIL import Image
import cv2
import numpy as np

import pandas as pd
```

```python


machine = ""
base = f"{machine}/media/z/0/MVPC10/DATA/v1.1/RAW"
device = "device_03"
outpath = f"{base}/{device}"
csvfile = f"{outpath}/output.csv"

df = pd.read_csv(f'{csvfile}')
df.head
```

```python
df1 = df['image_name']
df1.head
```

```python
# for i in range(len(df1)):
#     print(df1.iloc[i])

li = []
cnt = 0
previous = ""
for index, row in df1.iteritems():
    current = row
    if previous == current:
        cnt += 1
    else:
        li.append([previous, cnt])
        cnt = 0
        previous = current
    previous = ""
li
```

```python

df2 = pd.DataFrame(li, columns=['img_name', 'cnt'])
df2
```

```python

df2.to_csv(f"{outpath}/v1.2_output.csv")
```
